##### <u>Sep parameter</u>
- This is the seperator parameter. If its a tsv file then while reading the file you will have to mention `pd.read_csv("train.csv", sep = "\t")`

##### <u>Name parameter</u>
- If there is no column names already in the data set then you can use this parameter as `pd.read_csv("train.csv", name=["id", "name", "year", "rating"])`

##### <u>Index_col parameter</u>
-  Make a column from the data set as the index column instead of the default one. `pd.read_csv("train.csv", index_col="year")`

```python
df = pd.read_csv("aug_train.csv", index_col="city")
df
```

##### <u>use_cols parameter</u>
- To use only specified columns from the data set rather than having all the columns. `pd.read_csv("train.csv", usecols=["id", "name"])`

```python
pd.read_csv("aug_train.csv", usecols=["city", "gender"])
```

##### <u>skiprows/nrows parameter</u>
- `pd.read_csv("train.csv", skiprows=[1,5])`
- `pd.read_csv("train.csv", nrows=[6])`

```python
pd.read_csv("aug_train.csv", nrows=3)
```

##### <u>dtypes parameter</u>
- Change the data type of a column

```python
pd.read_csv("aug_train.csv", dtype={"target": int})
```

##### <u>Handling dates</u>
- Date values are by default treated as object(string) in pandas
- `pd.read_csv("train.csv", parse_dates=["column_name"])`

##### <u>Convertors</u>
- To convert the values of an column

```python
def rename(name):
    if (name == "no_enrollment"):
        return "NaN"
    else:
        return name

pd.read_csv("aug_train.csv", converters={"enrolled_university": rename})
```

##### <u>na_values parameter</u>
- To change things like `-` to `NaN`
- `pd.read_csv("train.csv", na_values=["-", "something"])`

##### <u>Loading huge data in chunks</u>

```python
dfs = pd.read_csv("aug_train.csv", chunksize=5000)

for chunks in dfs:
    print(chunks.shape)
```
