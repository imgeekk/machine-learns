### Applying gradient descent
#### creating a class for convenience

```python
class GDRegressor:
    def __init__(self, learning_rate, epochs):
        self.m = 100    # random initial slope value
        self.b = -120   # random initial b value
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X, y):
        # calculating new b and m values
        for i in range(self.epochs):
            loss_slope_wrt_b = -2 * np.sum(y - self.m*X.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slope_wrt_b)
            loss_slope_wrt_m = -2 * np.sum((y - self.m*X.ravel() - self.b) * X.ravel())
            self.m = self.m - (self.lr * loss_slope_wrt_m)
        print("m =", self.m, "b =", self.b)
```

```python
gd = GDRegressor(0.001, 100)
```

```python
gd.fit(X, y)
```

### Look we converged to the actual $m$ and $b$ values
