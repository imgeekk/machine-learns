```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

#### Creating our own dataset using make_regression()

```python
X, y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)
```

```python
pd.DataFrame(X).head()
```

```python
pd.DataFrame(y).head()
```

```python
df = pd.DataFrame({"feature1": X[:, 0], "feature2": X[:, -1], "target": y})

df.head()
```

#### we have created a dataset which is having 100 rows and 3 columnn - 2 inputs and 1 output

```python
px.scatter_3d(df, x="feature1", y="feature2", z="target")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=3
)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("mae =", mean_absolute_error(y_test, y_pred))
print("mse =", mean_squared_error(y_test, y_pred))
print("r2 score =", r2_score(y_test, y_pred))
```

#### Visualising the regression plane

```python
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
xGrid, yGrid = np.meshgrid(y, x)

final = np.vstack((xGrid.ravel().reshape(1, 100), yGrid.ravel().reshape(1, 100))).T

z_final = lr.predict(final).reshape(10, 10)

z = z_final
```

```python
fig = px.scatter_3d(df, x="feature1", y="feature2", z="target")

fig.add_trace(go.Surface(x = x, y = y, z = z))

fig.show()
```
