```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
import seaborn as sns
plt.figure(figsize=(5, 3))
sns.scatterplot(df, x="cgpa", y="package")
plt.show()
```

```python
X = df.iloc[:, [0]]
y = df.iloc[:, [1]]

# OR

# X = df.iloc[:, 0:1]
# y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=2
)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.predict(X_test.iloc[2].values.reshape(1,1))
```

#### Lets look at the line created by linear regression

```python
sns.scatterplot(df, x="cgpa", y="package")
plt.plot(X_train, lr.predict(X_train), color="red")

plt.show()
```

### m and b

```python
m = lr.coef_
b = lr.intercept_
```

```python
# line equation
# y = mx + b

# y is the output i.e package
# x is the cgpa
# So linear regression under the hood first finds the m and b and then keeps on predicting y/package with the given x/cgpa
```

```python
y = m * 8.58 + b # cgpa = 8.58
print(y)

print(lr.predict(X_test[X_test["cgpa"] == 8.58].values.reshape(1,1)))
```

#### Look they are the same
