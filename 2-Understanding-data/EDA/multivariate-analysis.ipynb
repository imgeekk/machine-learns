```python
titanic = pd.read_csv("../train.csv")
```

```python
tips = sns.load_dataset("tips")
```

```python
flights = sns.load_dataset("flights")
```

##### 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# to change color of the plot based on some other col
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker", size="size")
```


##### 1. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(data=titanic, x="Pclass", y="Age")
```

```python
sns.barplot(data=titanic, x="Pclass", y="Age", hue="Sex")
```

##### 1. Box Plot (Numerical - Categorical)

```python
sns.boxplot(data=titanic, x="Age", y="Sex")
```

```python
sns.boxplot(data=titanic, x="Age", y="Sex", hue="Survived")
```

##### 4. Distplot (Numerical - Categorical)

```python
# distplot is deprecated

# sns.displot(data=titanic[titanic["Survived"] == 0], x="Age", kind="kde")
# sns.displot(data=titanic[titanic["Survived"] == 1], x="Age", kind="kde")

# can also directly use kdeplot for kde (kde means the curve)
sns.kdeplot(data=titanic[titanic["Survived"] == 0], x="Age")
sns.kdeplot(data=titanic[titanic["Survived"] == 1], x="Age")
```

##### 5. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic["Pclass"], titanic["Survived"]))
```

```python
titanic.groupby("Pclass")["Survived"].mean()*100
```

##### 6. ClusterMap (Categorical - Categorical) 

```python
pd.crosstab(titanic["Pclass"], titanic["Survived"])
```

```python
sns.clustermap(pd.crosstab(titanic["Pclass"], titanic["Survived"]))
```
