## 1. On Cols having categorical data

### a. <u>Countplot</u>

```python
import seaborn as sns
```

```python

# sns.countplot(df["Survived"])
# This will not work because Survived is practically a numeric data of 0 1. Thats why seaborn will consider it as continous data

sns.countplot(x="Survived", data=df)
# â¬† This is to be done only when the col numeric but meant to be categorical. If the col already has string/object values, seaborn automatically treats them as categories. So you will just have to write -
# sns.countplot(df["Sex"])
```

```python
df["Survived"].value_counts().plot(kind="bar")
```

### b. <u>PieChart</u>

```python
df["Pclass"].value_counts().plot(kind="pie", autopct="%.7f")
```
