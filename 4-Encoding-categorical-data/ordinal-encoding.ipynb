## Ordinal Encoding

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("customer.csv")
```

```python
df.sample(5)
```

Here</br>
Age - Nig this is not categorical data, its numerical</br>
Gender - No order ergo Nominal</br>
Review - Has order ergo Ordinal</br>
Education - Has order ergo Ordinal</br>
Purchased - Has order ergo Ordinal
</br></br>
We are doing ordinal encoding in this file so consider Review, Education and Purchased columns. But note we will be doing <u>ordinal encoding in Review and Educated</u> ***but*** <u>Label encoding in Purchased column</u>

```python
# df = pd.read_csv("customer.csv", usecols=[2, 3, 4])
# df.head()
df = df.iloc[:, 2:]
# [left, right] left part decides how many rows you want, and right part tells how many cols you want
# so df.iloc[2:5, 1:3] will give me rows 2 to 4 with columns 1 to 2

df.head()
```

Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :2],
                                                    df.iloc[:, [2]],
                                                    test_size=0.2)
X_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[["Poor", "Average", "Good"], ["School", "UG", "PG"]])

oe.fit(X_train)
X_train_encoded = oe.transform(X_train)
X_test_encoded = oe.transform(X_test)
```

```python
# Now converting the resultant numpy array
X_train_encoded = pd.DataFrame(X_train_encoded, columns=X_train.columns)
X_train_encoded.head()
```

```python
oe.categories_
```

# Label Encoding
Only used to encode target values i.e y and not input x
Here column Purchased which we have already seperated to y_train and y_test

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(y_train)

y_train_encoded = le.transform(y_train)
y_test_encoded = le.transform(y_test)
```

```python
le.classes_  # same as oe.categories_
```

```python
y_train_encoded = pd.DataFrame(y_train_encoded, columns=y_train.columns)
y_train_encoded.head()
```
